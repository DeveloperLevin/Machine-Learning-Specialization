```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# read CSV file
df = pd.read_csv('../data/areas.csv')
df
```

```python
# plot area vs price
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color="red", marker='+')
```

# Training the Model

```python
# Creating a Linear Regression object
X = df.iloc[:,:1]
y = df.iloc[:,-1]
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

# Prediction

```python
# predict price for area of 3000
result = reg.predict([[3000]])
print(f"The predicted price is: {result}")

print(f"The Slope is: {reg.coef_}") # w0 - slope
print(f"The Intercept is: {reg.intercept_}") # b - intercept
```

# Plot Prediction

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df['area'], y, color='blue', marker='*', label='Data Points')
plt.plot(df['area'], reg.predict(df[['area']]), color='red', linewidth=2, label='Regression Line')
plt.legend()
plt.show()
```
